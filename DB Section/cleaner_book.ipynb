```python
import pandas as pd
import numpy as np
from ast import literal_eval
```

```python
df_main = pd.read_csv('books.csv')
df_main.info()
```

```python
df_edited = df_main.drop_duplicates('ids_book').drop(['Unnamed: 0', 'Unnamed: 36'], axis= 1)
df_edited.info()
```

<h3>Fixing Publish Year</h3>

```python
df_edited['sunny_date'] = df_edited['sunny_date'].str.extract(r'(\d+)', expand=False)
df_edited['sunny_date'] = pd.to_numeric(df_edited['sunny_date'], errors='coerce').fillna(np.nan)
df_edited['monthy_date'] = df_edited['monthy_date'].str.extract(r'(\d+)', expand=False)
df_edited['monthy_date'] = pd.to_numeric(df_edited['monthy_date'], errors='coerce').fillna(np.nan)
df_edited['monthy_date'].info()
```

### Fixing Another Title

```python
df_edited['shabak'] = df_edited['shabak'].str.extract(r'(\d+-?\d+)', expand=False)
df_edited['shabak'] = df_edited['shabak'].fillna(np.nan)
df_edited['shabak'].info()
```

### Pages and Pices

```python
pages = df_edited['pages'].str.extract('(\d+)', expand=False)
pages = pd.to_numeric(pages, errors='coerce').fillna(np.nan)
df_edited['pages'] = np.where(pages.notnull(), pages, df_edited['pages'])
df_edited['pages'] = df_edited['pages'].astype(float)
df_edited['pages'].info()
```

### Series

```python
df_edited['seriess'] = pd.to_numeric(df_edited['seriess'], errors='coerce').fillna(np.nan)
df_edited['seriess'].info()
```

### Itration Values

```python
iter_val = df_edited[['writer_name', 'writer_id', 'writer_link', 'translations', 'ids_translations', 'book_tags', 'ids_tags', 'english_dialogs', 'persian_dialogs', 'authors', 'pleasure_text']]
```

```python
iter_val = iter_val.replace(np.nan, "[None]")
iter_val.head()
```

```python
for col in iter_val.columns:
    iter_val[col] = iter_val[col].apply(literal_eval)
iter_val.head()
```

```python
df_edited[['writer_name', 'writer_id', 'writer_link', 'translations', 'ids_translations', 'book_tags', 'ids_tags', 'english_dialogs', 'persian_dialogs', 'authors', 'pleasure_text']] = iter_val
df_edited.head()
```

## Finalizing

```python
df_edited = df_edited.drop(['publisher_link', 'writer_link', 'link_translations'], axis= 1)
df_edited.to_csv('Final_Data.csv', index=False, encoding='utf-8-sig')
```
