```python
import pandas as pd
import numpy as np
```

```python
bt_df = pd.read_csv('Book_Tag.csv')
books = pd.read_csv('All_bools.csv')
bt_df.head()
```

```python
books.info()
```

```python
tags_df = pd.read_csv('Tags.csv')
tags_df.head()
```

```python
id_love = tags_df[tags_df['book_tags'] == 'عاشقانه']['ids_tags']
id_love = id_love.to_list()[0]
print(id_love)
```

```python
new_df = books.merge(bt_df, on=['ids_book'])
new_df
```

```python
love_books = new_df[new_df['ids_tags'] == id_love]
love_books
```

```python
love_books = love_books.sort_values(['seriess', 'rate'], ascending= False).iloc[:20, ]
love_books
```

```python
writers = pd.read_csv('Writer.csv')
bw_df = pd.read_csv('Book_writer.csv')
```

```python
miane = bw_df.merge(writers, on= 'writer_id')
miane.head()
```

```python
fi_book = miane[miane['ids_book'].isin(love_books['ids_book'])]
# fi_book = miane[miane['ids_book'] in love_books['ids_book']]
fi_book
```

Factors for choosing is Rate by books

```python
names = fi_book.drop_duplicates('writer_name')
names = names['writer_name'].to_list()[:5]
print(names)
```
